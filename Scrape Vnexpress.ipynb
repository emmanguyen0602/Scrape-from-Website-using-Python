```python
import numpy as np
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

from time import sleep
import random
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException, ElementClickInterceptedException
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Declare browser
s = Service('C://Users//hantt3//Downloads//chromedriver_win32')
driver = webdriver.Chrome(service=s)
url = 'https://timkiem.vnexpress.net/?q=vi%E1%BB%87n%20d%C6%B0%E1%BB%A1ng%20l%C3%A3o'

# Open URL
driver.get(url)
```

```python
elems = driver.find_elements(By.CSS_SELECTOR , '.title-news [href]')
links = [elem.get_attribute('href') for elem in elems]
```

```python
def getComment(link):
    driver.get(link)
    content_comment, like_count = [], []
    try: #this code only click xem them y kien 1 lan
        next_pagination_cmt = driver.find_element(By.XPATH, "/html/body/section[7]/div/div[3]/div[1]/div[2]/div/div/div[8]/a")
        next_pagination_cmt.click()
        print("Xem thêm ý kiến")
        sleep(5)
            
    except ElementClickInterceptedException:
        print("ElementClickInterceptedException!")
        
    except NoSuchElementException:
        print("NoSuchElementException!")
        
    except ElementNotInteractableException:
        print ("ElementNotInteractableException")
    elems_content = driver.find_elements(By.CSS_SELECTOR , ".full_content")
    content_comment = [elem.text for elem in elems_content] + content_comment
            
    elems_likeCount = driver.find_elements(By.CSS_SELECTOR, '.reactions-total .number')
    like_count = [elem.text for elem in elems_likeCount] + like_count

        
    df = pd.DataFrame(list(zip(content_comment, like_count)), 
                       columns = ['content_comment','like_count'])
    df.insert(0, "link_item", link)
    return df
```

```python
df_list = []
for link in links:
    df = getComment(link)
    df_list.append(df)
```
