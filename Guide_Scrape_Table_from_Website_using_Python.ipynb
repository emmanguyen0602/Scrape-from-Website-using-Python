#STEP 2. IMPORT REQUIRED LIBRARIES


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#STEP 3. SELECT PAGE

#STEP 4. REQUEST PERMISSION

```python
# Create an URL object
url = 'https://www.worldometers.info/coronavirus/'
# Create object page
page = requests.get(url)
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
```

#STEP 5. INSPECT TABLE ELEMENT

```python
#inspect the location of the table
# Obtain information from tag <table>
table1 = soup.find("table", id="main_table_countries_today")
```

#STEP 6. CREATE A COLUMN LIST

```python
#Inspect location of columns
# Obtain every title of columns with tag <th>
headers = []
for i in table1.find_all("th"):
 title = i.text
 headers.append(title)
headers
```

```python
headers[13]="Tests/1M pop"
```

#STEP 7. CREATE A DATA FRAME

```python
# Create a dataframe
mydata = pd.DataFrame(columns = headers)
```

#STEP 8. CREATE A FOR LOOP TO FILL DATAFRAME

```python
#identify the location of the row and item column
#the row is located under tag <tr> and items are located under tag <td> 

# Create a for loop to fill mydata
for j in table1.find_all("tr")[1:]:
 row_data = j.find_all("td")
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```

```python
mydata
```

#STEP 9. CLEANING THE DATAFRAME

```python
# Drop and clearing unnecessary rows
mydata.drop(mydata.index[0:7], inplace=True)
mydata.drop(mydata.index[222:229], inplace=True)
mydata.reset_index(inplace=True, drop=True)
# Drop “#” column
mydata.drop("#", inplace=True, axis=1)
```
